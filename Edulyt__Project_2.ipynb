```python
df = df.merge(df1[['C_ID', 'Gender',]], on='C_ID', how='left')
```

```python
#converting age into grps according gender and age
def age_grouper(age,gender):
  if age<=35 and gender=='M':
    return 'Young Males'
  elif 35 < age <=60 and gender == 'M':
        return 'Mid age Males'
  elif age >60 and gender == 'M':
        return 'Old Males'
  elif age <= 35 and gender == 'F':
        return 'Young Females'
  elif 35 < age <=60 and gender == 'F':
        return 'Mid age Females'
  elif age >60 and gender == 'F':
        return 'Old Females'
df['Age_group'] = df.apply(lambda row: age_grouper(row['Age'], row['Gender']), axis=1)
```

```python
#Now we are calculating how many swipes does each segment did
swipes=df.groupby('Age_group').size().reset_index(name='NO-OF-Swipes')
swipes=swipes.sort_values(by='NO-OF-Swipes',ascending=False).reset_index(drop=True)
```

```python
#These are the number of swipes in each segment as per above function
swipes
```

```python
spend=df.groupby(['Age_group'],as_index=False)['Selling_price'].sum()
spend=spend.sort_values(by='Selling_price',ascending=False).reset_index(drop=True)
```

```python
#These are the spends of each segment based on age and gender i have segmented in above function
spend
```

```python
def time_grouper(time):
  a=time.hour
  if a>=0 and a<=6:
    return 'Midnight'
  elif a>6  and a<=12:
    return 'Morning'
  elif a>12 and a<=18:
    return 'Afternoon'
  else:
    return 'Night'

df['timing']=df.apply(lambda time: time_grouper(time['Time']),axis=1)
```

```python
df.head()
```

```python
#converting age into grps
def age_grp(age):
  if age<=35 :
    return 'Young'
  elif 35 < age <=60 :
        return 'Mid age '
  elif age >60 :
        return 'Old'
df['Age_seg'] = df.apply(lambda row: age_grp(row['Age']), axis=1)
```
