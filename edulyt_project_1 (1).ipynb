# **Imposing interest to customers who has dues after billing time **

```python
reward_df['diff_S&R']=reward_df['Amount']-reward_df['Amount_repay']
```

```python
reward_df
```

```python
#reward_df[reward_df['diff_S&R']<=0]=0
```

```python
reward_df.loc[reward_df['diff_S&R']<=0,'diff_S&R']=0
```

```python
reward_df
```

```python
reward_df=reward_df.rename(columns={'diff_S&R':'dues'})
```

```python
new_df = reward_df[['Costomer', 'Year','month','Penalty']]
```

```python
new_df
```

```python
new_df.to_excel('bank profit.xlsx', index=False)
```

**imposing penalty to customers who has dues and penalty is also referred as profit for the bank from that customer**

*   **Penalty is 2.9% of dues **



```python
reward_df['Penalty']=((2.9/100)*reward_df['dues'])
```

##  Which is the most profitable segment?

*   which segment has generating more penalty that is the most profitable segment so we will sumup all peanalties of segments




```python
most_profitable_segment=reward_df.groupby(['Segment'])['Penalty'].sum().reset_index()
```

```python
most_profitable_segment
```

```python
most_profitable_segment=most_profitable_segment.sort_values(by='Penalty',ascending=False)
```

```python
most_profitable_segment
```

From this we can conclude that from normal people profit is generating more
